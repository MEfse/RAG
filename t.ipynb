```python
def mergesortedarray(nums1, m, nums2, n):
    i = m - 1
    j = n - 1
    k = m + n - 1

    while j >= 0:
        if nums1[i] > nums2[j] and i >= 0:
            nums1[k] = nums1[i]
            i -= 1
        else:
            nums1[k] = nums2[j]
            j -= 1

        k -= 1

    return nums1



print(mergesortedarray([1,2,3,0,0,0], 3, [2,5,6], 3))
```

```python
def validparentheses(s):
    stack = []
    mapping = {')':'(', '}':'{', ']':'['}

    for char in s:
        if char in mapping.values():
            stack.append(char)
        elif char in mapping:
            if not stack or stack[-1] != mapping[char]:
                return False
            stack.pop()

    return not stack
            
            



print(validparentheses("()[]{}"))
print(validparentheses("(]"))
print(validparentheses("([{}])"))
```

```python
def search(nums):
    if len(nums) == 0:
        return 0
    
    nums.sort()

    longest = 1
    current = 1

    for i in range(1, len(nums)):
        if nums[i] == nums[i - 1]:
            continue
            
        elif nums[i] == nums[i - 1] + 1:
            current += 1

        else:
            longest = max(longest, current)
            current = 1

    return longest

search([100,4,200,1,3,2])
```

```python
def search(nums):
    nums_set = set(nums)

    longest = 0

    for num in nums_set:
        if num - 1 not in nums_set:
            current_num = num 
            current_streak = 1

            while current_num + 1 in nums_set:
                current_num += 1 
                current_streak += 1

            longest = max(longest, current_streak)

    return longest

search([100,4,200,1,3,2])
```

```python
from pydantic import BaseModel

class IngredientInput(BaseModel):
    ingredient_text: str


class ParsedIngredient(BaseModel):
    ingredient: str
    quantity: float
    unit: str


import re

UNIT_MAP = {
    "стакан": 200,  # ml
    "л": 1000,
    "мл": 1
}

from transformers import pipeline

ner_model = pipeline("token-classification", model="bert-base-cased")


def extract_with_ner(text: str):
    entities = ner_model(text)

    ingredient = None
    quantity = None
    unit = None

    for e in entities:
        if e["entity"] == "QUANTITY":
            quantity = float(e["word"])
        elif e["entity"] == "UNIT":
            unit = e["word"]
        elif e["entity"] == "FOOD":
            ingredient = e["word"]

    return {
        "ingredient": ingredient,
        "quantity": quantity,
        "unit": unit
    }

def normalize(parsed: dict):
    unit = parsed["unit"]
    quantity = parsed["quantity"]

    if unit in UNIT_MAP:
        return {
            "ingredient": parsed["ingredient"],
            "quantity": quantity * UNIT_MAP[unit],  # ✅ остаётся quantity
            "unit": "ml"
        }

    return parsed

def process_ingredient(text: str):
    # 1. extraction (regex or ML)
    parsed = extract_with_ner(text)

    # 2. normalization
    normalized = normalize(parsed)

    return ParsedIngredient(**normalized)
```

```python
def subarray_sum(nums, k):
    prefix_count = {0 : 1}
    curr_sum = 0
    count = 0

    for num in nums:
        curr_sum += num 
        need = curr_sum - k

        if need in prefix_count:
            count += prefix_count[need]

        prefix_count[curr_sum] = prefix_count.get(curr_sum, 0) + 1

    return count


print(subarray_sum([1,1,1], 2))      # 2
print(subarray_sum([1,2,3], 3))      # 2
print(subarray_sum([1,-1,0], 0))     # 3
```

```python
def productofarrayacceptself(nums):
    n = len(nums)
    answer = [1] * n
    
    prefix = 1 
    for i in range(n):
        answer[i] = prefix
        prefix *= nums[i]

    suffix = 1 
    for i in range(n - 1, -1, -1):
        answer[i] *= suffix
        suffix *= nums[i]

    
    return answer


productofarrayacceptself([1,2,3,4])
```

```python
def max_profit(prices: list[int]) -> int:
    min_price = float('inf')
    max_profit = 0

    for price in prices:
        min_price = min(min_price, price)
        profit = price - min_price
        max_profit = max(max_profit, profit)

    return max_profit


# Примеры:
print(max_profit([7,1,5,3,6,4]))  # 5
print(max_profit([7,6,4,3,1]))    # 0
```

```python
def move_zeroes(nums: list[int]) -> list[int]:
    left = 0 

    for right in range(len(nums)):
        if nums[right] != 0:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1

    return nums


# Примеры:
print(move_zeroes([0,1,0,3,12]))
print(move_zeroes([0]))
print(move_zeroes([1,2,3]))
```

```python
def transaction_stats(transactions):

    hashmap = {}

    for transaction in transactions:
        user_id = transaction['user_id']
        amount = transaction['amount']

        if user_id not in hashmap:
            hashmap[user_id] = {
                'count': 0,
                'total': 0
            }

        hashmap[user_id]['count'] += 1
        hashmap[user_id]['total'] += amount


    for user_id in hashmap:
        total = hashmap[user_id]['total']
        count = hashmap[user_id]['count']

        hashmap[user_id]['avg'] = total / count

    return hashmap


transactions = [
    {"user_id": 1, "amount": 100},
    {"user_id": 2, "amount": 50},
    {"user_id": 1, "amount": 200},
    {"user_id": 3, "amount": 70},
    {"user_id": 2, "amount": 30},
]

print(transaction_stats(transactions))
```

```python
def is_palindrome(s: str) -> bool:
    s = ''.join(c.lower() for c in s if c.isalnum())

    left, right = 0, len(s) - 1

    while left < right:
        if s[left] == s[right]:
            left += 1
            right -= 1
        else:
            return False
        
    return True


print(is_palindrome("A man, a plan, a canal: Panama"))  # True
print(is_palindrome("race a car"))                      # False
print(is_palindrome(" "))                               # True
print(is_palindrome("0P"))                              # False
```

```python
from typing import List

def two_sum_sorted(numbers: List[int], target: int) -> List[int]:
    left, right = 0, len(numbers) - 1

    while left < right:  
        if numbers[left] + numbers[right] == target:
            return [left + 1, right + 1] 
        elif numbers[left] + numbers[right] < target:
            left += 1
        else: 
            right -= 1

    #return 

# Тесты:
print(two_sum_sorted([2,7,11,15], 9))           # [1,2]
print(two_sum_sorted([1,2,3,4,4,9,56,90], 8))   # [4,5]
print(two_sum_sorted([1,3,4,5,6], 10))          # [4,5]
```

```python
def length_of_longest_substring(s: str) -> int:
    left = 0 
    unique_char = set()
    max_length = 0 

    for right in range(len(s)):
        while s[right] in unique_char:
            unique_char.remove(s[left])
            left += 1
        unique_char.add(s[right])
        max_length = max(max_length, right - left + 1)
        
    return max_length
        


# Примеры:
print(length_of_longest_substring("abcabcbb"))  # 3 ("abc")
print(length_of_longest_substring("bbbbb"))     # 1 ("b")
print(length_of_longest_substring("pwwkew"))    # 3 ("wke")
print(length_of_longest_substring(""))          # 0
```

```python
from typing import List

def min_subarray_len(target: int, nums: List[int]) -> int:
    left = 0
    min_lenght = float('inf')
    sum_nums = 0

    for right in range(len(nums)):
        sum_nums += nums[right]

        while sum_nums >= target:
            min_lenght = min(min_lenght, right - left + 1)
            sum_nums -= nums[left]
            left += 1
            
    if min_lenght == float('inf'):
        return 0
                           
    return min_lenght 
        



print(min_subarray_len(7, [2,3,1,2,4,3]))  # 2
print(min_subarray_len(4, [1,4,4]))        # 1
print(min_subarray_len(11, [1,1,1,1,1]))   # 0
```

```python
def check_inclusion(s1: str, s2: str) -> bool:
    if len(s1) > len(s2):
        return False
    
    s1_count = {}
    s2_count = {}

    for i in range(len(s1)):
        s1_count[s1[i]] = 1 + s1_count.get(s1[i], 0)
        s2_count[s2[i]] = 1 + s2_count.get(s2[i], 0)

    if s1_count == s2_count:
        return True
    
    left = 0
    for right in range(len(s1), len(s2)):
        s2_count[s2[right]] = 1 + s2_count.get(s2[right], 0)
        s2_count[s2[left]] -= 1

        if s2_count[s2[left]] == 0:
            del s2_count[s2[left]]

        left += 1

        if s1_count == s2_count:
            return True
        
    return False



print(check_inclusion("ab", "eidbaooo"))  # True
print(check_inclusion("ab", "eidboaoo"))  # False
```

```python
from typing import List
from collections import defaultdict

def group_anagrams(strs: List[str]) -> List[List[str]]:
    hashmap = {}

    for word in strs:
        key = ''.join(sorted(word))
        if key not in hashmap:
            hashmap[key] = []
        hashmap[key].append(word)

    return list(hashmap.values())

print(group_anagrams(["eat","tea","tan","ate","nat","bat"]))
print(group_anagrams([""]))
print(group_anagrams(["a"]))
```

```python
from typing import List

def search_insert(nums: List[int], target: int) -> int:
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return left 

print(search_insert([1,3,5,6], 5))  # 2
print(search_insert([1,3,5,6], 2))  # 1
print(search_insert([1,3,5,6], 7))  # 4
print(search_insert([1,3,5,6], 0))  # 0
```

```python
from typing import List

def search_range(nums: List[int], target: int) -> List[int]:
    def find_left():
        left, right = 0, len(nums) - 1
        index = -1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                index = mid
                right = mid - 1  # продолжаем искать слева
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
        return index

    def find_right():
        left, right = 0, len(nums) - 1
        index = -1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                index = mid
                left = mid + 1  # продолжаем искать слева
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
        return index

    return [find_left(), find_right()]

print(search_range([5,7,7,8,8,10], 8))
print(search_range([5,7,7,8,8,10], 6))
print(search_range([], 0))
```
