```python
import requests

response = requests.post(
    "http://localhost:8001/embed",
    json={
        "text": "How to print Hello World in Python?"
    }
)

print(response.status_code)
print(response.text)
```

```python
import requests

response = requests.post(
    "http://localhost:8010/predict",
    json={
        "pairs": [
            ["How to print Hello World in Python?", "Use print('Hello World')"]
        ]
    }
)

print(response.status_code)
print(response.text)
```

```python
import requests

response = requests.post(
    "http://localhost:8020/generate",
    json={"prompt": "How to print Hello World in Python?"}
)

print(response.status_code)
print(response.text)
```
